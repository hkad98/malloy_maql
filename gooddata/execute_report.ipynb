```python
from gooddata_pandas import GoodPandas
from gooddata_sdk import ObjId
import os
```

```python
gdf = GoodPandas(
    host=os.environ["TIGER_ENDPOINT"],
    token=os.environ["TIGER_API_TOKEN"]
).data_frames(workspace_id="faa_custom")
```

```python
indexed_df = gdf.indexed(
    index_by=dict(
        name=ObjId(id="nickname", type="label"),
    ),
    columns=dict(
        flight_count=ObjId(id="flight_count", type="metric"),
        destination_count=ObjId(id="destination_count", type="metric"),
        percentage_of_flights=ObjId(id="percentage_of_flights", type="metric"),
    ),
).sort_values("percentage_of_flights", ascending=False)

display(indexed_df)
```

```python
indexed_df = gdf.for_insight(insight_id="flights_by_carrier").sort_values("percentage_of_flights", ascending=False)

display(indexed_df)
```
